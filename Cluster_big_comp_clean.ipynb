```python
best = []
for i in data.columns:
    if i == 'clustering':
        best.append('Best')
    elif (i == 'n_of_clusters') | (i == 'modularity'):
        for j in range(len(data[i])):
            d = data.loc[j, i]
            if d == '-': d = 0
            if d == max(data[data[i] != '-'][i]): b = data.loc[j, 'clustering']
        best.append(b)
    else:
        for j in range(len(data[i])):
            d = data.loc[j, i]
            if d == '-': d = 999999
            if d == min(data[data[i] != '-'][i]): b = data.loc[j, 'clustering']
        best.append(b)

data = data.append(pd.Series(best, index=data.columns), ignore_index=True)
```

```python
import FA_mesurev3
from FA_mesurev3 import FA_mesure_calc
```

```python
Counter(Counter(list(df_to_check['cluster_real'])).values())
```

```python
black_list = set()
C = Counter(list(df_to_check['cluster_real']))
for i in range(len(df_to_check)):
    cl_real = df_to_check.loc[i, 'cluster_real']
    if (C[cl_real] == 1) | (C[cl_real] == 80376):
        black_list.add(df_to_check.loc[i, 'inn'])
        
len(black_list)
```

```python
white_list = set(df_to_check.inn) - black_list
```

```python
def FA_measure_results(df_to_check, partition, beta_, emphasis_, white_list):
    test = df_to_check[df_to_check['inn'].isin(white_list)].copy()
    parts = partition
    partition_white_list = set(partition.keys()).intersection(white_list)
    partition = {key: parts[key] for key in partition_white_list}
    test['pred'] = test.inn.map(partition)

    test1 = test.loc[~test.pred.isna()]
    test1['pred'] = test1['pred'].astype(int)

    pred_clusters = test1.pred.unique()
    partition_values = np.array(list(partition.values()))
    partition_values = partition_values[np.isin(partition_values, pred_clusters.astype(int))]

    act_clusters = test1.cluster_real.unique()
    act_values = np.array(list(df_to_check.cluster_real.values))
    act_values = act_values[np.isin(act_values, act_clusters)].astype('object')

    rids_decode, cids_decode, score, res_recall, res_precision = FA_mesurev3.FA_mesure_calc(test1.cluster_real.values, 
                                                                                            test1.pred.values, 
                                                                                            partition_values,
                                                                                            act_values,
                                                                                            beta=beta_,
                                                                                            emphasis = emphasis_)
    
    return rids_decode, cids_decode, score, res_recall, res_precision
```
